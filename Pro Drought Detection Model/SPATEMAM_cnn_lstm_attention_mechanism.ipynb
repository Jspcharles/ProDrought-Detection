```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, LeakyReLU, LSTM, TimeDistributed, Input, Attention, LayerNormalization, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# Directories
DATA_DIR = r"data\processed_v02"
MODEL_SAVE_PATH = r"model\cnn_lstm_am\cnn_lstm_attention_model.h5"

# Load preprocessed data
def load_data():
    X_train = np.load(os.path.join(DATA_DIR, "X_train.npy"))
    Y_train = np.load(os.path.join(DATA_DIR, "Y_train.npy"))
    X_test = np.load(os.path.join(DATA_DIR, "X_test.npy"))
    Y_test = np.load(os.path.join(DATA_DIR, "Y_test.npy"))

    # Ensure normalization
    X_train = np.clip(X_train, 0, 1)
    X_test = np.clip(X_test, 0, 1)

    print("Data loaded and normalized.")
    print(f"Training set: {X_train.shape}, Test set: {X_test.shape}")
    return X_train, Y_train, X_test, Y_test

# Build the Enhanced CNN + Bidirectional LSTM + Attention Model
def build_cnn_lstm_attention_model(input_shape, time_steps=10):
    # CNN Feature Extractor
    cnn_input = Input(shape=(32, 32, 5))
    x = Conv2D(32, (3, 3), activation='relu')(cnn_input)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)

    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)

    x = Flatten()(x)
    cnn_model = Model(inputs=cnn_input, outputs=x)

    # LSTM Model with Attention
    lstm_input = Input(shape=(time_steps, 32, 32, 5))
    x = TimeDistributed(cnn_model)(lstm_input)
    x = Bidirectional(LSTM(64, return_sequences=True))(x)
    x = Attention()([x, x])  # Self-Attention
    x = LayerNormalization()(x)
    x = LSTM(64)(x)
    x = Dropout(0.5)(x)
    x = Dense(128, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=lstm_input, outputs=output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model

# Training the Enhanced Model with Class Weights
def train_cnn_lstm_attention():
    X_train, Y_train, X_test, Y_test = load_data()

    # Configurable time steps
    time_steps = 10

    # Trim data to be divisible by time steps
    train_trim = len(X_train) - (len(X_train) % time_steps)
    test_trim = len(X_test) - (len(X_test) % time_steps)

    X_train = X_train[:train_trim]
    Y_train = Y_train[:train_trim]

    X_test = X_test[:test_trim]
    Y_test = Y_test[:test_trim]

    # Reshape for LSTM
    X_train = X_train.reshape((-1, time_steps, 32, 32, 5))
    Y_train = Y_train.reshape((-1, time_steps))[:, -1]

    X_test = X_test.reshape((-1, time_steps, 32, 32, 5))
    Y_test = Y_test.reshape((-1, time_steps))[:, -1]

    print(f"Reshaped Training Data: {X_train.shape}, Testing Data: {X_test.shape}")

    model = build_cnn_lstm_attention_model(X_train.shape[2:])

    # Calculate Class Weights
    class_weights = {0: 1.0, 1: 5.0}  # Emphasize minority class

    # Callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    checkpoint = ModelCheckpoint(MODEL_SAVE_PATH, monitor='val_loss', save_best_only=True)

    history = model.fit(
        X_train, Y_train,
        validation_data=(X_test, Y_test),
        class_weight=class_weights,
        epochs=30,
        batch_size=16,
        callbacks=[early_stopping, checkpoint],
        verbose=1
    )

    # Save final model
    model.save(MODEL_SAVE_PATH)

    # Training Visualization
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.show()

if __name__ == "__main__":
    train_cnn_lstm_attention()
```

```python
import os
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Directories
DATA_DIR = r"data\processed_v02"
MODEL_PATH = r"model\cnn_lstm_am\cnn_lstm_attention_model.h5"

# Load Model and Data
def load_model_and_data():
    model = tf.keras.models.load_model(MODEL_PATH)

    X_test = np.load(os.path.join(DATA_DIR, "X_test.npy"))
    Y_test = np.load(os.path.join(DATA_DIR, "Y_test.npy"))

    # Ensure data is properly shaped for LSTM
    time_steps = 10
    test_trim = len(X_test) - (len(X_test) % time_steps)

    X_test = X_test[:test_trim].reshape((-1, time_steps, 32, 32, 5))
    Y_test = Y_test[:test_trim].reshape((-1, time_steps))[:, -1]

    print(f"Test Data Shape: {X_test.shape}, Test Labels Shape: {Y_test.shape}")

    return model, X_test, Y_test

# Evaluate Model with Advanced Metrics
def evaluate_model():
    model, X_test, Y_test = load_model_and_data()

    # Predicting on Test Data
    predictions = model.predict(X_test)
    predictions = (predictions > 0.5).astype(int).flatten()

    # Classification Report
    print("\nClassification Report:\n")
    print(classification_report(Y_test, predictions, digits=4))

    # Confusion Matrix
    cm = confusion_matrix(Y_test, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Drought (0)", "Consecutive Drought (1)"])

    # Display Confusion Matrix
    plt.figure(figsize=(6, 6))
    disp.plot(cmap=plt.cm.Blues)
    plt.title("Confusion Matrix: CNN + LSTM Model")
    plt.show()

if __name__ == "__main__":
    evaluate_model()
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tqdm import tqdm

# === CONFIG ===
SEQUENCE_DIR = r"data/sequences/CHC"  # Change region here
MODEL_PATH = r"models/cnn_lstm_att_lowmem.h5"
TIME_SLICE = 120  # Time window index to filter (last time step in sequence)

# === LOAD MODEL ===
model = tf.keras.models.load_model(MODEL_PATH, compile=False)
print("✅ Model loaded")

# === LOAD AND PREDICT ===
predictions = {}

for file in tqdm(os.listdir(SEQUENCE_DIR)):
    if file.endswith(".npy") and not file.endswith("_label.npy") and f"_t{TIME_SLICE}" in file:
        # Parse lat, lon from filename
        parts = file.split("_")
        lat = int(parts[1].replace("lat", ""))
        lon = int(parts[2].replace("lon", ""))

        # Load and predict
        try:
            sequence = np.load(os.path.join(SEQUENCE_DIR, file)).astype(np.float32)
            if sequence.shape != (10, 32, 32, 5):
                continue
            pred = model.predict(sequence[np.newaxis, ...])[0][0]  # Scalar
            predictions[(lat, lon)] = pred
        except:
            continue

# === BUILD GRID ===
lats = sorted(set(k[0] for k in predictions.keys()))
lons = sorted(set(k[1] for k in predictions.keys()))

lat_to_idx = {lat: i for i, lat in enumerate(lats)}
lon_to_idx = {lon: j for j, lon in enumerate(lons)}

grid = np.full((len(lats), len(lons)), np.nan)
for (lat, lon), val in predictions.items():
    i, j = lat_to_idx[lat], lon_to_idx[lon]
    grid[i, j] = val

# === PLOT ===
plt.figure(figsize=(8, 6))
c = plt.imshow(grid, cmap="Reds", origin="lower")
plt.colorbar(c, label="Predicted Consecutive Drought Probability")
plt.title(f"Predicted Drought Map – Region: CHC, Time Index: {TIME_SLICE}")
plt.xlabel("Longitude Index")
plt.ylabel("Latitude Index")
plt.show()
```
