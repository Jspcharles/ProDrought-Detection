```python
import os
import numpy as np
from sklearn.utils import shuffle

SEQUENCE_DIR = "data/sequences"

# Get all region folders and shuffle (same as training split)
all_dirs = sorted(os.listdir(SEQUENCE_DIR))
train_split = int(0.8 * len(all_dirs))
train_dirs = shuffle(all_dirs, random_state=42)[:train_split]

# Count class labels
label_counts = {0: 0, 1: 0}

for region in train_dirs:
    region_path = os.path.join(SEQUENCE_DIR, region)
    for file in os.listdir(region_path):
        if file.endswith("_label.npy"):
            label_path = os.path.join(region_path, file)
            try:
                label = np.load(label_path).item()
                if label in [0, 1]:
                    label_counts[int(label)] += 1
            except Exception as e:
                print(f"Skipping {label_path}: {e}")

# Report
total = sum(label_counts.values())
print("ðŸ”Ž Class Distribution in Training Set:")
print(f"  Class 0 (No Drought): {label_counts[0]}")
print(f"  Class 1 (Consecutive Drought): {label_counts[1]}")
print(f"  Total Samples: {total}")
print(f"  Class Ratio (1:0) = {label_counts[1]}:{label_counts[0]}  ({label_counts[1] / total:.2%} drought)")
```
