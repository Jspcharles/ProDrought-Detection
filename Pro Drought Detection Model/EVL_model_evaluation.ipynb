```python
from tensorflow.keras.models import load_model

model = load_model("models/best_model_A2.h5")
```

```python
from data_generator import SequenceDataGenerator, get_region_dirs
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Load validation/test directories
all_dirs = get_region_dirs("data/sequences")
all_dirs.sort()  # to keep order reproducible
test_dirs = all_dirs[-int(0.2 * len(all_dirs)):]  # last 20% as test

test_gen = SequenceDataGenerator(test_dirs, batch_size=16, shuffle=False)
```

```python
y_true = []
y_pred = []

for X_batch, y_batch in test_gen:
    preds = model.predict(X_batch)
    y_true.extend(y_batch)
    y_pred.extend(preds.flatten())

y_pred_binary = [1 if p >= 0.5 else 0 for p in y_pred]

print("ðŸ“Š Classification Report (threshold=0.5):")
print(classification_report(y_true, y_pred_binary, digits=4))

print("ðŸ“‰ Confusion Matrix:")
print(confusion_matrix(y_true, y_pred_binary))

print(f"ðŸ”¥ ROC-AUC: {roc_auc_score(y_true, y_pred):.4f}")
```
