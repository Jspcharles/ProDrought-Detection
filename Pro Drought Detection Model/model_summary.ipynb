```python
from tensorflow.keras.models import load_model

# List of model filenames
model_files = [
    r"models\test_models\best_model_m1.h5",
    r"models\test_models\best_model_m2.h5",
    r"models\test_models\best_model_m3.h5",
    r"models\test_models\best_model_m4.h5",
    r"models\test_models\best_model_m5.h5",
    r"models\test_models\best_model_m6.h5"
]

# Dictionary to store loaded models
models = {}

# Load all models
for i, file in enumerate(model_files, start=1):
    models[f"Model_{i}"] = load_model(file)

# Compare models by printing summary
for name, model in models.items():
    print(f"\n{name} summary:")
    model.summary()
```
