```python

import matplotlib.pyplot as plt

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(12, 12))

# Plot the GeoDataFrame
gdf.plot(
    ax=ax,
    column='q_reg_name',
    cmap='summer',
    legend=True,
    edgecolor='black',        # Adds border lines between regions
    linewidth=0.5,            # Thinner border lines
    alpha=0.9                 # Slight transparency
)

# Add a title
ax.set_title('Bioregions of Queensland', fontsize=16)

# Remove axis for cleaner look
ax.set_axis_off()

for idx, row in gdf.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, row['q_reg_name'], fontsize=8, ha='center', color='black')


# Adjust legend
leg = ax.get_legend()
if leg:
    leg.set_title('Region Name')
    leg._legend_box.align = "left"

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import os

# Base directory where the folders will be created
base_dir = "Figures\\SPI vs NDVI\\0.5"

# Ensure the base directory exists
os.makedirs(base_dir, exist_ok=True)

# List of region names
region_names = [
    "Brigalow Belt", "Channel Country", "Central Queensland Coast", 
    "Cape York Peninsula", "Desert Uplands", "Einasleigh Uplands", 
    "Gulf Plains", "Mitchell Grass Downs", "Mulga Lands", 
    "New England Tableland", "Northwest Highlands", 
    "Southeast Queensland", "Wet Tropics"
]

# Create directories for each region
for region in region_names:
    region_dir = os.path.join(base_dir, region.replace(" ", "_"))
    os.makedirs(region_dir, exist_ok=True)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np

# Folder containing multiple geojson files
folder_path = "QLD Bio Regions/QLD_BioRegions_Separated"

# Load and combine all GeoJSON files
gdfs = []
for file in os.listdir(folder_path):
    if file.endswith(".geojson"):
        gdf = gpd.read_file(os.path.join(folder_path, file))
        gdf["q_reg_name"] = os.path.splitext(file)[0]  # Use filename as region name
        gdfs.append(gdf)

# Combine all into a single GeoDataFrame
combined_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

# Add dummy values for choropleth coloring (replace with real SPI/NDVI if available)
combined_gdf["value"] = np.random.randint(10, 100, len(combined_gdf))

# Plot static choropleth
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
combined_gdf.plot(column="value", cmap="YlGnBu", legend=True, edgecolor="black", ax=ax)
plt.title("Static Choropleth Map from Multiple GeoJSONs")
plt.axis('off')
plt.tight_layout()
# plt.savefig("choropleth_static_map.png", dpi=300)
plt.show()
```
