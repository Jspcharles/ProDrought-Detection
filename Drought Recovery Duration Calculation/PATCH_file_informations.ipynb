```python
import os
import xarray as xr

# === Path Settings ===
labeled_folder = r"Data - Modelling\Region_Monthly_DB\3. Con Drought Labelled Files"  # Folder containing labeled region files
regions = ["BRB_Combined_Consecutive.nc", "CHC_Combined_Consecutive.nc", "CQC_Combined_Consecutive.nc",
            "CYP_Combined_Consecutive.nc", "DEU_Combined_Consecutive.nc", "EIU_Combined_Consecutive.nc",
            "GUP_Combined_Consecutive.nc", "MGD_Combined_Consecutive.nc", "MUL_Combined_Consecutive.nc",
            "NET_Combined_Consecutive.nc", "NWH_Combined_Consecutive.nc", "SEQ_Combined_Consecutive.nc",
            "WET_Combined_Consecutive.nc"]  # Add your region filenames here

print("✅ Region Grid Sizes:")

# === Check Grid Size for Each Region ===
for region_file in regions:
    region_path = os.path.join(labeled_folder, region_file)
    ds = xr.open_dataset(region_path)
    
    # Extract grid dimensions (latitude x longitude)
    lat_size = ds.dims['lat']
    lon_size = ds.dims['lon']
    
    print(f"{region_file}: {lat_size} (latitude) x {lon_size} (longitude)")
```

```python
import os
import geopandas as gpd

# === Path Settings ===
geojson_folder = r"Data\QLD Bio Regions\QLD_BioRegions_Separated"  # Folder containing GeoJSON files
geojson_files = ["Brigalow_Belt.geojson", "Channel_Country.geojson", "Central_Queensland_Coast.geojson", 
    "Cape_York_Peninsula.geojson", "Desert_Uplands.geojson", "Einasleigh_Uplands.geojson", 
    "Gulf_Plains.geojson", "Mitchell_Grass_Downs.geojson", "Mulga_Lands.geojson", 
    "New_England_Tableland.geojson", "Northwest_Highlands.geojson", 
    "Southeast_Queensland.geojson", "Wet_Tropics.geojson"]  # Add your filenames here

print("✅ GeoJSON Region Bounding Boxes:")

# === Check Bounding Box for Each GeoJSON File ===
for file in geojson_files:
    file_path = os.path.join(geojson_folder, file)
    gdf = gpd.read_file(file_path)

    bounds = gdf.total_bounds  # returns [minx, miny, maxx, maxy]
    min_lon, min_lat, max_lon, max_lat = bounds

    print(f"{file}:")
    print(f"  Latitude range: {min_lat} to {max_lat}")
    print(f"  Longitude range: {min_lon} to {max_lon}")
    print()
```
