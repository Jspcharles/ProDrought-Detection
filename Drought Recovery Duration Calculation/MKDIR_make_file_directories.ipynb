```python
import os

# Base directory where the folders will be created
base_dir = "Data\Data for SPI\Vapour Pressure Deficit\VPD Monthly"

# Ensure the base directory exists
os.makedirs(base_dir, exist_ok=True)

# List of region names
region_names = [
    "Brigalow Belt", "Channel Country", "Central Queensland Coast", 
    "Cape York Peninsula", "Desert Uplands", "Einasleigh Uplands", 
    "Gulf Plains", "Mitchell Grass Downs", "Mulga Lands", 
    "New England Tableland", "Northwest Highlands", 
    "Southeast Queensland", "Wet Tropics"
]

# Create directories for each region
for region in region_names:
    region_dir = os.path.join(base_dir, region.replace(" ", "_"))
    os.makedirs(region_dir, exist_ok=True)
```
