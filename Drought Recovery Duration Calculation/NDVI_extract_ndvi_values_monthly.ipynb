```python
import pandas as pd
df = pd.read_csv('Data/MODIS/NDVI csv/monthly_mean_ndvi_2001_2023.csv')
```

```python
import plotly.graph_objects as go

# Filter data for a specific region, e.g., "Brigalow Belt"
region_name = "Brigalow Belt"
region_data = df[df['region_name'] == region_name]

# Create a Candlestick Chart
fig = go.Figure(data=[go.Candlestick(
    x=region_data['year'].astype(str) + '-' + region_data['month'].astype(str),
    open=region_data['mean_ndvi'],
    high=region_data['mean_ndvi'],
    low=region_data['mean_ndvi'],
    close=region_data['mean_ndvi']
)])

fig.update_layout(
    title=f'Candlestick Chart for {region_name}',
    xaxis_title='Year-Month',
    yaxis_title='Mean NDVI',
    xaxis_rangeslider_visible=False
)

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

# Create a datetime column from year and month
region_data['Date'] = pd.to_datetime(region_data[['year', 'month']].assign(day=1))

# Set the plot style
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(20, 6))
sns.lineplot(data=region_data, x='Date', y='mean_ndvi', marker='o')

# Format x-axis to show only the year
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Add titles and labels
plt.title(f'NDVI Values Over Time for {region_name}', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean NDVI', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Create a 'Year-Month' column for better visualization
region_data['Year-Month'] = region_data['year'].astype(str) + '-' + region_data['month'].astype(str)

# Set the plot style
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=region_data, x='Year-Month', y='mean_ndvi', marker='o')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=8)

# Add titles and labels
plt.title(f'NDVI Values Over Time for {region_name}', fontsize=16)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Mean NDVI', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the original CSV
df = pd.read_csv("Data/MODIS/NDVI csv/All_region_monthly_mean_ndvi_2001_2023.csv")

# Assuming your region column is named 'region'
# Group by the 'region' column and save each group as a separate CSV
for region_name, group in df.groupby("region_name"):
    # Clean the region name to be safe for filenames
    safe_region_name = str(region_name).replace(" ", "_").replace("/", "-")
    group.to_csv(f"Data/MODIS/NDVI csv/{safe_region_name}_ndvi.csv", index=False)
```

```python
import os
import pandas as pd

# Directory containing the CSV files
directory = "Data\Merged SPI & NDVI & VPD\Region wise CSV"

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):  # Process only region-specific NDVI files
        file_path = os.path.join(directory, filename)
        
        # Load the CSV file
        region_df = pd.read_csv(file_path)
        
        # Calculate minimum and maximum NDVI
        min_ndvi = region_df['NDVI'].min()
        max_ndvi = region_df['NDVI'].max()
        
        # Print the results
        print(f"Region: {filename.replace('.csv', '').replace('_', ' ')}")
        print(f"Minimum NDVI: {min_ndvi}")
        print(f"Maximum NDVI: {max_ndvi}")
        print("-" * 40)
```
