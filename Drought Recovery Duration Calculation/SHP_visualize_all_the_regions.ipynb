```python
import pandas as pd
import geopandas as gpd
import glob

import matplotlib.pyplot as plt

# Path to the folder containing your GeoJSON files
geojson_folder = "Data\QLD Bio Regions\QLD_BioRegions_Separated"

# Load all GeoJSON files into a single GeoDataFrame
geojson_files = glob.glob(f"{geojson_folder}/*.geojson")
bioregions = [gpd.read_file(file) for file in geojson_files]
combined_gdf = gpd.GeoDataFrame(pd.concat(bioregions, ignore_index=True))

# Plot the bioregions with distinct colors
fig, ax = plt.subplots(figsize=(12, 10))
combined_gdf.plot(ax=ax, column='q_reg_name', cmap='tab20', legend=True)

# Add title and labels
ax.set_title("Bioregions of Queensland", fontsize=16)
ax.set_xlabel("Longitude", fontsize=12)
ax.set_ylabel("Latitude", fontsize=12)

# Save the plot as an image for your academic paper
plt.savefig("queensland_bioregions.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
for region_gdf in bioregions:
    region_name = region_gdf['q_reg_name'].iloc[0]  # Get the region name
    fig, ax = plt.subplots(figsize=(8, 6))
    region_gdf.plot(ax=ax, color='lightblue', edgecolor='black')
    ax.set_title(f"Region: {region_name}", fontsize=14)
    ax.set_xlabel("Longitude", fontsize=12)
    ax.set_ylabel("Latitude", fontsize=12)
    plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import glob
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

geojson_folder = "Data\\QLD Bio Regions\\QLD_BioRegions_Separated"

geojson_files = glob.glob(f"{geojson_folder}/*.geojson")
bioregions = [gpd.read_file(file) for file in geojson_files]
combined_gdf = gpd.GeoDataFrame(pd.concat(bioregions, ignore_index=True))

fig, ax = plt.subplots(figsize=(12, 10), dpi=300)

combined_gdf.plot(
    ax=ax,
    column='q_reg_name',
    cmap='Blues',
    legend=True,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={
        'fontsize': 11,        #  legend text size
        'title_fontsize': 10   #  legend title size
    }
)

# Axis format (E / S)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{abs(x):.0f}E"))
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, pos: f"{abs(y):.0f}S"))

# Increase tick label font size
ax.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import glob
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

geojson_folder = "Data\\QLD Bio Regions\\QLD_BioRegions_Separated"

geojson_files = glob.glob(f"{geojson_folder}/*.geojson")
bioregions = [gpd.read_file(file) for file in geojson_files]
combined_gdf = gpd.GeoDataFrame(pd.concat(bioregions, ignore_index=True))

# Create numbered legend labels
unique_regions = sorted(combined_gdf['q_reg_name'].unique())
region_map = {
    name: f"{name} ({i+1:02d})"
    for i, name in enumerate(unique_regions)
}
combined_gdf['q_reg_label'] = combined_gdf['q_reg_name'].map(region_map)

fig, ax = plt.subplots(figsize=(12, 12), dpi=300)

combined_gdf.plot(
    ax=ax,
    column='q_reg_label',
    cmap='Blues',
    legend=True,
    edgecolor='black',
    linewidth=0.5,
    legend_kwds={
        'fontsize': 11,        #  legend text size
        'title_fontsize': 10   #  legend title size
    }
)

# Axis format (E / S)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{abs(x):.0f}E"))
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, pos: f"{abs(y):.0f}S"))

# Increase tick label font size
ax.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```
